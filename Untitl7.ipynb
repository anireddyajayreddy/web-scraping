```
from bs4 import BeautifulSoup
import requests
```

```
url = 'https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_4_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=378c3a4f-7d6b-44a0-a7c9-bdc81008ec68&as-backfill=on'
page = requests.get(url)
```

```
page.text
```

```
soup = BeautifulSoup(page.text,'html.parser')
```

```

soup.title
```

```
name = []
price = []
for a in soup.findAll('a',href=True, attrs={'class':'_31qSD5'}): #css class using inspect, using findAll method
  ni = a.find('div', attrs={'class':'_3wU53n'})
  pi = a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
  name.append(ni.text)
  price.append(pi.text)
```

```
 #creating data frame 
import pandas as pd

df = pd.DataFrame(list(zip(name,price)))
df.columns = ['Name','Price']
```
